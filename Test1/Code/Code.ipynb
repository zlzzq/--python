## 3 计算词频

```python
#合并成一个文档
text = []
for word in titles3:
    text.extend(word)
```

```python
#统计频数
from collections import defaultdict
word_dict = defaultdict(int) 
for word in text:
    word_dict[word] +=1
```

```python
word_dict.items()
```

```python
#从字典中提取词条和词频
terms = []
freq = []
for item in word_dict.items():
    terms.append(item[0])
    freq.append(item[1])
```

```python
wordcount = pd.DataFrame({'terms':terms,'count':freq})#写入数据框
wordcount1 = wordcount.sort_values(by='count',ascending=False)#按词频降序排列
wordcount1
```

## TF-IDF

```python
#定义用于tfidf的分词器
def tokench(text):
    tokens = [word for word in text.split(' ')]#数据以空格隔开
    after_tokens = []
    for token in tokens:
        if len(token)>1:
            after_tokens.append(token)
    return after_tokens
```

```python
#构建tfidf矩阵
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                 tokenizer=tokench,use_idf=True)

tfidf_matrix = tfidf_vectorizer.fit_transform(df['titlecut']) #计算任何两个文档的相似度
```

```python
terms = tfidf_vectorizer.get_feature_names()#获取词（特征）名
terms[:10]
```

```python
weight = tfidf_matrix.todense().tolist()#tfidf权重
```

## 文本相似度

```python
#计算距离
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(tfidf_matrix)
dist = 1.0 - cosine_similarity(tfidf_matrix)#根据两个文档的词条的TF-IDF值来计算
```

```python
import pandas as pd
# df = pd.read_csv('train_content.csv', error_bad_lines=False,header=None)
df = pd.read_csv('train_content.csv',encoding='utf-8')
```

```python
df = df[['Data']]
df.dropna(inplace=True)
df
```

## nltk分词

```python
import nltk
texts = []
for text in df['Data']:
    tokens = [token.lower() for token in nltk.word_tokenize(text)]
    tagged = nltk.pos_tag(tokens)
    tagged1 = [[word[0],word[1]] for word in tagged]
    texts.append(tagged1)
```

```python
texts[1]
```

```python
##按词性提取
texts1 = [[word[0] for word in tagged if word[1]=='JJ' or word[1]=='JJR' or word[1]=='JJS'
        or word[1]=='FW' or word[1]=='NN' or word[1]=='NNS' or word[1]=='NNP'
        or word[1]=='NNPS' or word[1]=='RB' or word[1]=='RBR' or word[1]=='RBS'
        or word[1]=='RP' or word[1]=='VB' or word[1]=='VBD' or word[1]=='VBG'
        or word[1]=='VBN' or word[1]=='VBP' or word[1]=='VBZ'] for tagged in texts]
```

```python
texts1[1]
```

## 过滤stopwords


```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
texts2 = [[word for word in text if word not in stopwords] for text in texts1]
#以空格分割
texts3 = [' '.join(text) for text in texts2]
df['descut'] = texts3
df
```

```python
#定义用于tfidf的分词器
def tokench(text):
    tokens = [word for word in text.split(' ')]#数据以空格隔开
    after_tokens = []
    for token in tokens:
        if len(token)>1:
            after_tokens.append(token)
    return after_tokens
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.8, max_features=200000,
                                 tokenizer=tokench,use_idf=True)

tfidf_matrix = tfidf_vectorizer.fit_transform(df['descut'])
```

```python
#计算距离矩阵
from sklearn.metrics.pairwise import cosine_similarity
dist = 1 - cosine_similarity(tfidf_matrix)
```

## 主题建模

```python
from gensim import corpora, models, similarities
texts = titles3
```

```python
dictionary = corpora.Dictionary(texts) #构造词典
dictionary.filter_extremes(no_below=1, no_above=0.8)#去高频词

corpus = [dictionary.doc2bow(text) for text in texts] #构造语料库
```

```python
[s for s in corpus]
```

```python
#训练一个LDA模型
%time lda = models.LdaModel(corpus, num_topics=9, id2word=dictionary, update_every=5, alpha=0.1, eta=0.1)
```

```python
print(lda[corpus[0]])#打印第一个文档的主题建模结果
```

**预测文档对应的主题**

```python
doctop = [ s for s in lda[corpus]]
doctop = pd.DataFrame(doctop)
doctop
```

```python
doctop.to_csv('DocumentTopic.csv')
```

**主题对应的关键词**

```python
topicwords = lda.print_topics(num_words=20)
topicwords
```
